```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_hdf("vict05to19.h5", "vict")
df
```

```python
print(df.columns.values)
```

```python
data = df[['DT_DAY',
           'TX_DAY_OF_WEEK_DESCR_NL',
           'MS_VICT',
           'MS_SLY_INJ',
           'MS_DEAD_30_DAYS',
           'TX_VICT_TYPE_DESCR_NL',
           'TX_ROAD_USR_TYPE_DESCR_NL',
          'TX_ROAD_TYPE_DESCR_NL',
          'TX_AGE_CLS_DESCR_NL',
          'TX_PROV_DESCR_NL',
          'TX_RGN_DESCR_NL'
          ]]
data
```

```python
data = data.rename({"DT_DAY":"Year", 
                    "TX_DAY_OF_WEEK_DESCR_NL":"Day of week",
                    "MS_VICT":"Number of victims",
                    "MS_SLY_INJ":"Seriously injured",
                    "MS_DEAD_30_DAYS":"Dead within 30 days",
                    "TX_VICT_TYPE_DESCR_NL":"Victim type",
                    "TX_ROAD_USR_TYPE_DESCR_NL":"Vehicle",
                    "TX_ROAD_TYPE_DESCR_NL":"Road type",
                    "TX_AGE_CLS_DESCR_NL":"Age",
                    "TX_PROV_DESCR_NL":"Province",
                    "TX_RGN_DESCR_NL":"Region"},
                   axis=1)
data.head()
```

```python
datayear = data.groupby(by=[data["Year"].dt.strftime("%Y"), "Region"]).sum()
datayear
```

```python
datayear2 = datayear.reset_index()
```

```python
datayear2["Region"] = datayear2["Region"].replace("Brussels Hoofdstedelijk Gewest", "Brussels Capital Region")
datayear2["Region"] = datayear2["Region"].replace("Vlaams Gewest", "Flemish Region")
datayear2["Region"] = datayear2["Region"].replace("Waals Gewest", "Walloon Region")
datayear2
```

```python
plt.figure(figsize=(30,15))
sns.set_context("poster")
sns.barplot(x="Year", y="Number of victims", hue="Region", data=datayear2)
plt.savefig("firstplot.png", dpi=400, facecolor="w")
```

```python
plt.figure(figsize=(30,15))
sns.set_context("poster")
sns.lineplot(x="Year", y="Number of victims", hue="Region", data=datayear2)
```

```python
datayear2.to_excel("datayear2.xlsx")
```
